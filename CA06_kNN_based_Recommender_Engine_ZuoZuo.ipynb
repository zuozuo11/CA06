```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

**1 Data Source and Contents**

```python
data = pd.read_csv('https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv') #import data
```

```python
display(data.head())
display(data.describe())
display(data.info())
display(data.isnull().sum())
```

```python
data.drop(columns = 'Label', inplace=True) # From the instruction we know column 'label' is useless, so drop down this column
```

```python
display(data.head())
display(data.columns) # Show dataset after cleaning
```

```python


#feature slection
#split dataset in features and target variable
feature_cols = ['IMDB Rating', 'Biography', 'Drama', 'Thriller', 'Comedy', 'Crime', 'Mystery', 'History']
X = data[feature_cols]# Feature
Y = data[['Movie Name']] #target variable
```

```python
# Train the model using the dataset
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X, Y)
```

```python
#IMDB Rating = 7.2, Biography = Yes, Drama = Yes, Thriller = No, Comedy = No, Crime = No,Mystery = No, History = Yes
#Substitute the data of 'the post' to calculate the distance to other movies
Target = classifier.kneighbors([[7.2, 1, 1, 0, 0, 0, 0, 1]], return_distance= False)
```

```python
# show the results of above recommandation
for item in Target:
    rec = data.iloc[item]['Movie Name']
   
print(f'Here are movies you might like based on your interest in "the post": \n{rec}')
```
